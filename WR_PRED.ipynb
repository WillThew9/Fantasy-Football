```python
import sqlite3
import pandas as pd
pd.set_option('display.max_columns', None)
import plotly.express as px
```

```python
# Import my stats table from my data base to a pandas dataframe
conn = sqlite3.connect('NFL_STATS.db')
query = "SELECT * FROM WR_STATS_TABLE"
df = pd.read_sql_query(query, conn)
conn.close()
df
```

```python
# Check for NA values - There are none
df.isna().sum()
```

```python
# Check for data types
df.dtypes
```

```python
# Remove special characters from certain rows that were causing trouble
df['COMP_PCT'] = df['COMP_PCT'].str.replace('%', '')
df['REC_YDS'] = df['REC_YDS'].str.replace(',', '')
df['SNAP_COUNT'] = df['SNAP_COUNT'].str.replace(',', '')
```

```python
# Change data types that need to be changed
cols_to_change = df[["COMP_PCT", "REC_YDS", "SNAP_COUNT"]]

for col in cols_to_change:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.dtypes
```

```python
# Ensure no data lost
df.count()
```

```python
# Get summary statistics
df.describe()
```

```python
# Check the sums of rows that say they have nothing in them except for 0
potential_drops = df[['OVER_THREE_HUNDRED', 'RUSH_HUNDRED_YDS']]
potential_drops.sum()
```

```python
# Drop said columns
df = df.drop(['OVER_THREE_HUNDRED', 'RUSH_HUNDRED_YDS'], axis=1)
df.describe()
```

```python
# Checked on who was playing more than 17 games, I am chalking this up as a typo because it is the only one in the data set
df[df['GP'] > 17]
```

```python
# Reset his GP value to the number of games that were played that season
df.at[630, 'GP'] = 16
```

```python
df.columns
```

```python
# I don't care how the receiver got the yards, so I made TOTAL_YDS
df['TOTAL_YDS'] = df['PASS_YDS'] + df['RUSH_YDS'] + df['REC_YDS']
df.describe()
```

```python
# Ranked each receiver by a medley of statistics comparing them to each other
df = df.sort_values(by=['POINTS', 'REC', 'REC_YDS', 'SNAP_COUNT', 'TGT'], ascending=[False, False, False, False, False])

df['DEPTH'] = df.groupby(['YEAR', 'TEAM']).cumcount() + 1
df['DEPTH'] = df['DEPTH'].apply(lambda x: f'WR{x}')
df
```

```python
# Filtered out guys who don't score many points, they may be irrelevant in fantasy football
df = df[df['POINTS_G'] > 5.0]
df
```

```python
# Checking how many receivers were included per team, weird to see a WR5 in there
fig = px.histogram(df, x='TEAM', color='DEPTH', title='WRs Per Team')
fig.show()
```
